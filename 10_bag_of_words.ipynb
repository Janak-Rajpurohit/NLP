```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("spam.csv")
df.head()
```

```python
df["Category"].value_counts()
```

```python
df["spam"] = df.Category.apply(lambda x: 1 if x=='spam' else 0)
df.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test  = train_test_split(df["Message"],df['spam'],test_size=0.2)
len(x_train)
```

```python
type(x_train)
```

```python
x_train[:5]
```

```python
from sklearn.feature_extraction.text import CountVectorizer

v = CountVectorizer()
# .values convert the series into numpy array
x_train_cv = v.fit_transform(x_train.values)
x_train_cv
```

```python
x_train_cv.toarray()[:2]
```

```python
x_train_cv.shape
# 7726 unique words
```

```python
# getting the words in vocabulary
v.get_feature_names_out()[1000:2000]
```

```python
dir(v)
```

```python
# getting the number assing to words in vocublary
v.vocabulary_
```

```python
x_train_np = x_train_cv.toarray()
x_train_np[0]
```

```python
np.where(x_train_np[0]!=0)
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(x_train_cv,y_train)
```

```python
x_test_cv = v.transform(x_test)
```

```python
from sklearn.metrics import classification_report
y_pred = model.predict(x_test_cv)

print(classification_report(y_test,y_pred))
```

```python
emails = [
    "Hello brother, how are you?, Today i am arriving at airport.",
    "50% off on Amazon Great Indian Sale!! Click on the link below for discount."
]

email_cv = v.transform(emails)
model.predict(email_cv)
```

```python
# above all things can be done through sklearn pipeline
# it does everthing automatically first through count vectorizer then model multimonial nb
from sklearn.pipeline import Pipeline
clf = Pipeline([
    ('vectorizer',CountVectorizer()),
    ('nb',MultinomialNB())
])
```

```python
clf.fit(x_train,y_train)
```

```python
print(classification_report(y_test,clf.predict(x_test)))
```
